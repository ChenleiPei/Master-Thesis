```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import torch

latent_vectors = np.load('results/latent_space_parametric_3.npz')
# check the demension
print(latent_vectors.shape)

latent_tensor = torch.from_numpy(latent_vectors)
latent_mean, latent_log_std  = torch.unbind(latent_tensor, dim=-1)                                
print(latent_mean.shape)  
print(latent_log_std.shape)   
```

```python
import matplotlib.pyplot as plt
import torch  

plt.figure(figsize=(12, 10))  
plt.imshow(latent_mean, aspect='auto', cmap='viridis')  
plt.colorbar()  
plt.title('Heatmap of Tensor Values')
plt.xlabel('Feature Index')
plt.ylabel('Sample Index')
plt.show()
```

```python
plt.figure(figsize=(12, 10))  # 设置图像大小
plt.imshow(latent_log_std, aspect='auto', cmap='viridis')  
plt.colorbar()  
plt.title('Heatmap of Tensor Values')
plt.xlabel('Feature Index')
plt.ylabel('Sample Index')
plt.show()
```

```python
data = latent_mean.numpy()  


mean_values = np.mean(data, axis=0)  
std_values = np.std(data, axis=0)  
min_values = np.min(data, axis=0)  
max_values = np.max(data, axis=0)  
median_values = np.median(data, axis=0)  
percentile_25 = np.percentile(data, 25, axis=0)  
percentile_75 = np.percentile(data, 75, axis=0)  


stats_df = pd.DataFrame({
    'Mean': mean_values,
    'Std Dev': std_values,
    'Min': min_values,
    'Max': max_values,
    'Median': median_values,
    '25th Percentile': percentile_25,
    '75th Percentile': percentile_75
})


print(stats_df)

```
